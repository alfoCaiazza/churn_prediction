```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
DF_PATH = '../../data/dataframes/telecom_churn.csv'
df = pd.read_csv(DF_PATH)
df.head()
```

```python
df['Churn'].value_counts()
```

```python
X = df.drop(columns=['Churn', 'ContractRenewal', 'DataPlan'])
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(df_scaled, columns=X.columns, index=X.index)

# Insert removed columns
df_final = pd.concat([df_scaled, df[['Churn', 'ContractRenewal', 'DataPlan']]], axis=1)
df_final.head()
```

```python
df_final.to_csv('../../data/dataframes/telecom_churn_scaled.csv', index=True)
```
